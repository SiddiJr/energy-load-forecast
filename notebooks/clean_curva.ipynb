```python
import pandas as pd
```

```python
curva_2023 = pd.read_csv('data/raw/CURVA_CARGA_2023.csv', delimiter=';')
```

```python
def clean_curva(df):
    df = df[df['nom_subsistema'] == 'SUDESTE']
    df = df.drop(columns = ['id_subsistema', 'nom_subsistema'])
    df = df.rename(columns = {'din_instante': 'date', 'val_cargaenergiahomwmed': 'energy_demand'})
    df['date'] = pd.to_datetime(df['date'])
    df['time'] = df['date'].dt.time
    df['date'] = pd.to_datetime(df['date'].dt.date)
    df['is_weekend'] = df['date'].dt.dayofweek > 4
    return df
```

```python
curva_2023 = clean_curva(curva_2023)
curva_2023
```

```python
curva_2024 = pd.read_csv('data/raw/CURVA_CARGA_2024.csv', delimiter=';')
curva_2024 = clean_curva(curva_2024)
curva_2024
```

```python
curva = pd.concat([curva_2023, curva_2024])
curva.to_csv('data/clean/curva_horaria.csv')
```
